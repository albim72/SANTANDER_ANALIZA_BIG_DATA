```python
import pandas as pd
import re
from collections import Counter

# -----------------------------
# 1. Wczytanie logów
# -----------------------------
with open("server_big.log", "r") as f:
    lines = f.readlines()

df = pd.DataFrame({"raw": lines})

# -----------------------------
# 2. Parsowanie wyrażeniem regularnym
#    Format:
#    [timestamp] LEVEL message
# -----------------------------
pattern = r"\[(.*?)\]\s+(INFO|WARN|ERROR)\s+(.*)"

df[["timestamp", "level", "message"]] = df["raw"].str.extract(pattern)

# Konwersja timestampu
df["timestamp"] = pd.to_datetime(df["timestamp"])

# -----------------------------
# 3. Policz wystąpienia poziomów logów
# -----------------------------
counts = df["level"].value_counts()
print("Liczba wystąpień poziomów logów:")
print(counts)
print()

# -----------------------------
# 4. Wyświetl tylko logi typu ERROR
# -----------------------------
errors = df[df["level"] == "ERROR"]
print("ERROR logs:")
print(errors)
print()

# -----------------------------
# 5. Najczęściej występujące słowa
# -----------------------------
all_words = " ".join(df["message"].dropna()).lower()
all_words = re.findall(r"[a-zA-Z]+", all_words)

word_freq = Counter(all_words).most_common(10)
print("Najczęściej występujące słowa:")
print(word_freq)
```
